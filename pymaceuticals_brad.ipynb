```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
drugs_df = pd.read_csv(mouse_drug_data_to_load)
trials_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
drugs_trials_df = pd.merge (drugs_df, trials_df, on="Mouse ID")

# Display the data table for preview
print (drugs_df.describe())
print (trials_df.describe())
print (drugs_trials_df.describe())
```

```python
# THERE IS A PROBLEM IN THE DATA

# There are 1893 rows in the "trials_df".  I'm expecting that when I merge the "drugs_df", I will end up with the
# same number of rows.  However, there are 1906 rows in the merged "drugs_trials_df".  

# Also, in the "describe" results for the "drugs_df", there are 250 "Mouse IDs" but only 249 are unique.  There is
# one mouse who is listed twice.

# For that one mouse, since I can't determine which "trials_df" data goes with which "drug", I will have to remove that
# one mouse from the dataset.

# Fortunately, the "describe" results for "drugs_df" notes that the "top" "Mouse ID", or most frequently appearing 
# "Mouse ID" is "g989", and it appears with a "freq" of 2.  That is the rouge mouse!  Delete all references to 
# "Mouse ID = g989" in both datasets, and re-merge the data.

drugs_df2 = drugs_df.loc[(drugs_df["Mouse ID"] != "g989")]
trials_df2 = trials_df.loc[(trials_df["Mouse ID"] != "g989")]
drugs_trials_df2 = pd.merge (drugs_df2, trials_df2, on="Mouse ID")

print (drugs_df2.describe())
print (trials_df2.describe())
print (drugs_trials_df2.describe())
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
drugs_trials_gb = drugs_trials_df2.groupby (['Drug','Timepoint'])
mean_drugs_trials_df = drugs_trials_gb.mean()

# Convert to DataFrame
mean_drugs_trials_df2 = mean_drugs_trials_df.reset_index()
mean_drugs_trials_df2 = mean_drugs_trials_df2.drop(columns="Metastatic Sites")

# Preview DataFrame
mean_drugs_trials_df2.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
sem_drugs_trials_df = drugs_trials_gb.sem()

# Convert to DataFrame
sem_drugs_trials_df2 = sem_drugs_trials_df.reset_index()
sem_drugs_trials_df2 = sem_drugs_trials_df2.drop(columns=["Metastatic Sites","Mouse ID"])

# Preview DataFrame
sem_drugs_trials_df2.head()

```

```python
# Minor Data Munging to Re-Format the Data Frames
# Get the list of drugs
drugs = mean_drugs_trials_df2["Drug"].unique()

# Set a flag
first_pass = True

# For each of the drugs in the study
for drug in drugs:
    
    # Find the rows in the dataset that match that drug
    mean_drug_data = mean_drugs_trials_df2.loc[mean_drugs_trials_df2["Drug"]==drug]
    sem_drug_data = sem_drugs_trials_df2.loc[sem_drugs_trials_df2["Drug"]==drug]
    
    # Rename the "Tumor Volume (mm3)" column to be the name of the drug
    mean_drug_data.rename (columns={"Tumor Volume (mm3)":drug}, inplace=True)
    sem_drug_data.rename (columns={"Tumor Volume (mm3)":drug}, inplace=True)
    
    # Change the index to be the "Timepoint"
    mean_drug_data.set_index ("Timepoint", inplace=True)
    sem_drug_data.set_index ("Timepoint", inplace=True)
    
    # Delete the column "Drug"
    mean_drug_data.drop (columns="Drug", inplace=True)
    sem_drug_data.drop (columns="Drug", inplace=True)
    
    # IF this is the First Pass, then copy "drug_data" into "drugs_data", otherwise append the df to the larger df
    if first_pass:
        mean_drugs_data = mean_drug_data
        sem_drugs_data = sem_drug_data
        first_pass = False
    else:
        mean_drugs_data = pd.merge(mean_drugs_data, mean_drug_data, on="Timepoint", how="outer")
        sem_drugs_data = pd.merge(sem_drugs_data, sem_drug_data, on="Timepoint", how="outer")
    
# Preview that Reformatting worked
mean_drugs_data.head(20)
```

```python
# Generate the Plot (with Error Bars)

# Set the size of the graph
plt.figure(figsize=(10,10))

# Give it a title, and labels for the axes
plt.title("Tumor Response to Treatment", fontsize=16, fontweight="bold")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")

# Add grid lines
plt.grid()

# Create a list of marker types.  I couldn't figure out how to have "plt" show a default marker
# so I created this list.  I know the dataset only has 10 drugs, but I put more in in case a future run
# of the data has up to 15 drugs
marker_type = ['o','v','^','<','>','s','p','*','h','H','1','2','3','4','+']

# Draw the lines for each drug
for i, drug in enumerate(mean_drugs_data):
    plt.errorbar(mean_drugs_data.index, mean_drugs_data[drug], yerr=sem_drugs_data[drug],
                 linewidth=2, label=drug, marker=marker_type[i], markersize=7)

# plt.legend(handles=drugs, loc="best")
plt.legend (loc="best")

# Save the Figure
plt.savefig ('graph1.png')

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
# This line already ran in earlier cell, no need to run it again

# drugs_trials_gb = drugs_trials_df2.groupby (['Drug','Timepoint'])  
mouse_count_trials_df = drugs_trials_gb.size()

# Convert to DataFrame
mouse_count_trials_df2 = mouse_count_trials_df.reset_index()
mouse_count_trials_df2.rename (columns={0:"Mouse Count"}, inplace=True)

# Preview DataFrame
mouse_count_trials_df2.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

# Set a flag
first_pass = True

# For each of the drugs in the study
for drug in drugs:
    
    # Find the rows in the dataset that match that drug
    mouse_count_data = mouse_count_trials_df2.loc[mouse_count_trials_df2["Drug"]==drug]
    
    # Rename the "Metastatic Sites" column to be the name of the drug
    mouse_count_data.rename (columns={"Mouse Count":drug}, inplace=True)
    
    # Change the index to be the "Timepoint"
    mouse_count_data.set_index ("Timepoint", inplace=True)
    
    # Delete the column "Drug"
    mouse_count_data.drop (columns="Drug", inplace=True)
    
    # IF this is the First Pass, then copy "drug_data" into "drugs_data", otherwise append the df to the larger df
    if first_pass:
        mouse_count_data_all = mouse_count_data
        first_pass = False
    else:
        mouse_count_data_all = pd.merge(mouse_count_data_all, mouse_count_data, on="Timepoint", how="outer")
    
# Preview that Reformatting worked
mouse_count_data_all.head()
```

```python
# Calculate the percentages

# Make a copy of the mouse_count, call it mouse_pct
mouse_pct_data_all = mouse_count_data_all
for drug in drugs:
    start_mice = mouse_count_data_all[drug][0]
    mouse_pct_data_all[drug] = mouse_count_data_all[drug] / start_mice * 100

mouse_pct_data_all.head(20)
```

```python
# Generate the Plot (Accounting for percentages)

# Set the size of the graph
plt.figure(figsize=(10,10))

# Give it a title, and labels for the axes
plt.title("Survival during Treatment", fontsize=16, fontweight="bold")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")

# Add grid lines
plt.grid()

# Create a list of marker types.  I couldn't figure out how to have "plt" show a default marker
# so I created this list.  I know the dataset only has 10 drugs, but I put more in in case a future run
# of the data has up to 15 drugs

# Same list as before, so no reason to run this line again
# marker_type = ['o','v','^','<','>','s','p','*','h','H','1','2','3','4','+']

# Draw the lines for each drug
for i, drug in enumerate(mean_mr_drugs_data):
    plt.plot(mouse_pct_data_all.index, mouse_pct_data_all[drug], 
                 linewidth=2, label=drug, marker=marker_type[i], markersize=7)

# plt.legend(handles=drugs, loc="best")
plt.legend (loc="best")

# Save the Figure
plt.savefig ('graph3.png')

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
tumor_chgs = []
for col_num, drug in enumerate(drugs):
    tumor_chg2 = ( mean_drugs_data.iloc[9,col_num] / mean_drugs_data.iloc[0,col_num] -1 ) * 100
    tumor_chgs.append (tumor_chg2)
    
# Display the data to confirm
tumor_chgs
```

```python
# Store all Relevant Percent Changes into a Tuple
# I'm not sure why the comments suggest putting the data into a tuple vs dataframe, but ok, here it is.
tumor_tuple = tuple (tumor_chgs)
print (f"The tumor_tuple is {tumor_tuple}\n")

# Splice the data between passing and failing drugs
# Again, not sure why, but ok
passing_drugs = [drug for i, drug in enumerate(drugs) if tumor_chgs[i] < 0]
failing_drugs = [drug for i, drug in enumerate(drugs) if tumor_chgs[i] >= 0]
print (f"The list of passing drugs is {passing_drugs}")
print (f"The list of failing drugs is {failing_drugs}\n")

# What would be helpful is to have a list of colors for the bar chart
bar_colors = ['g' if tumor_chgs[i]<0 else 'r' for i, drug in enumerate(drugs) ]
print (f'The bar colors are {bar_colors}')

# Orient widths. Add labels, tick marks, etc.
fig, ax = plt.subplots(figsize=(12,12))

tick_locations = np.arange(len(drugs))
rects = ax.bar (tick_locations, height=tumor_chgs, align="edge", color=bar_colors)

ax.set_title ("Tumor Change Over 45 Day Treatment", weight="bold", size=20)
ax.set_ylabel ("% Tumor Volume Change", size=16)
ax.set_xlabel ("Drugs", size=16)

# ax.set_xticklabels(tick_locations, drugs, rotation=0, ha="left")
ax.set_xticklabels(drugs, horizontalalignment="left")
ax.set_xticks(tick_locations)

ax.grid ()

# Use functions to label the percentages of changes
def autolabel(rects, ax):
    # Get y-axis height to calculate label position from.
    (y_bottom, y_top) = ax.get_ylim()
    y_height = y_top - y_bottom

    for rect in rects:
        height = rect.get_height()

        # Fraction of axis height taken up by this rectangle
        p_height = (height / y_height)

#       Original code from the website http://composition.al/blog/2015/11/29/a-better-way-to-add-labels-to-bar-charts-with-matplotlib/
#          If we can fit the label above the column, do that;
#          otherwise, put it inside the column.
#        if p_height > 0.95: # arbitrary; 95% looked good to me.
#            label_position = height - (y_height * 0.05)
#        else:
#            label_position = height + (y_height * 0.01)

        # Position the label at +3 on the y-axis if the value of the bar is positive,
        # or set it to -3 if the value is negative, as those positions look good to me.
        if height > 0:
            label_position = 3
        else:
            label_position = -3
    
        # Write the text on the bars
        ax.text(rect.get_x() + rect.get_width()/2., label_position,
                '%.1f%%' % height, ha='center', va='bottom', color='white', size=12)

# Call functions to implement the function calls
autolabel(rects, ax)

# Save the Figure
plt.savefig ('graph4.png')

# Show the Figure
plt.show()
```

## Conclusions

1. I'm glad I expanded the study to include all 10 "drugs" given, as it appears that there are two that are equally effective, while 8 others don't work.

2. 3. When looking at "Metatastic Spread during Treatment", again Capomulin and Ramicane have the best data. Stelasyn comes in a close 3rd, with it's standard error bars overlapping with Capomulin.

5. Finally, in terms of the tumor volume change, Capomulin and Ramicane were the only ones to show declines, while the other 8 showed increases in size.
